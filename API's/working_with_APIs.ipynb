# API REQUESTS

We are using rickandmorty api which is an opne API with no authentication

```python
import requests

base_url = "https://rickandmortyapi.com/api/"

#there are various endpoints in this API which are character , location,and episode

endpoint = "character"
r = requests.get(base_url + endpoint)
print(r.json())
```

```python
data = r.json()
print(data["info"])  #info is one of the keys in the json file
```

```python
pages = print(data["info"]["pages"])

name = print(data["results"][0]["name"])
```

```python
episodes = data["results"][0]["episode"]
```

```python
#checking how many episodes the character is in

print(len(episodes))
```

```python
#A function to get the pages 

def get_pages(response):
  return response["info"]["pages"]

get_pages(data)#shows the number of pages in the data
```

```python
#creating a function to access many enpoints

def main_request(base_url,endpoint):
  r = requests.get(base_url + endpoint)
  return r.json()


data = main_request(base_url,endpoint)
print(data)
```

```python
#since there are many names of characters we want to create a function to be abl eto get the name and the episodes in which the character is in

def parse_json(response):
  charlist = []
  for item in response["results"]:
    char = {
      "id": item["id"],
      "name": item["name"],
      "no_of_episodes":len(item["episode"])}
    charlist.append(char)
  return charlist

parse_json(data)
```

```python
# Create a dataframe and be able to access different pages of the data to be able to get the necessary character information

import pandas as pd

def main_request_pages(base_url,endpoint,x):
  r = requests.get(base_url + endpoint + f"?page={x}")
  return r.json()

mainlist = []
data = main_request_pages(base_url,endpoint,1)
for x in range(1,get_pages(data)+1):
  mainlist.extend(parse_json(main_request_pages(base_url,endpoint,x)))
  
print(mainlist)
  
```

```python
#creating the dataframe csv file

df = pd.DataFrame(mainlist)

print(df.head(), "\n", df.tail())
```

```python
#saving the dataframe

df.to_csv("charlist.csv",index=False)
```
