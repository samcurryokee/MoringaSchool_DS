# Data Visualization

```python
import matplotlib.pyplot as plt

# Set plot space as inline for inline plots and qt for external plots
%matplotlib inline
```

## Scatter Plots

```python
import numpy as np

x = np.linspace(0,10,30)

y = np.sin(x)


plt.figure(figsize=(10,6))
plt.scatter(x,y , label = "Function sin(x)")
plt.title("Scatter Plot in Matplotlib")
plt.legend()
plt.xlabel("values of x")
plt.ylabel("sinx")
plt.show()
```

## Bar Graphs

```python
#generate some data 
np.random.seed(100)

# Generate the x-axis variable as 10 categories using numpy's arange function

x= np.arange(10)

#for y axis , generate 10 random quantities
y = np.random.rand(10)
```

```python
# Replace None with appropriate code
height = [40, 30, 10, 50, 25, 5]
x = range(len(height))
labels = ['Crime', 'Sci-Fi', 'Drama', 'Comedy', 'Action', 'Documentary']

# Create the plot
fig, ax = plt.subplots(figsize=(8, 6))

# Plot vertical bars of fixed width by passing x and height values to .bar() function 
ax.bar(x, height, tick_label=labels)

# Give a title to the bar graph and label the axes
ax.set_title("Jim's Video Library - Movie Genres")
ax.set_xlabel('Genres')
ax.set_ylabel('Number of Movies')

plt.show()
```

## Histograms
This is used to check the underlying frequency distributionof a dataset.
Italso alows you to visualize the fundamental properties about the data like if it is skewed in any particular direction or has outliers

```python
#Set seed 
np.random.seed(100)

#generate a data set of 200 retirement age values
x = 5* np.random.randn(200) + 65

#plot histogram
plt.hist(x, bins= 10, edgecolor ="black")

plt.xlabel("Retirement Age")
plt.ylabel("Frequency of Values")
plt.title("Histogram in Matplotlib")
plt.show()

```

```python
# Set seed for reproducability
np.random.seed(100)

# Generate a data set of 10000 retirement age values
x = 5*np.random.randn(10000) + 65

#Plot the distogram with hist() function
plt.hist(x, bins=50, edgecolor="black")

plt.xlabel('Retirement Age')
plt.ylabel('Frequency of Values')
plt.title('Histograms in Matplotlib')
plt.show()
```

## Customizing Vizualizations with Matplotlib


```python
#do the necessary imports
# Import pyplot for plotting
import matplotlib.pyplot as plt
# Import numpy to generate some dummy data
import numpy as np
%matplotlib inline
```

```python
years = range(1975,2000)

data = np.linspace(0,1000,25)
data
```

## Line Plots in Matplotlib

```python
fig,ax =plt.subplots()

ax.plot(years, data)

ax.legend(["Sample Data"],loc=4)
ax.set_xlabel("X Axis Label")
ax.set_ylabel("Y Axis Label")
ax.set_title("PLOT TITLE")
```

## Multiple Plots on the same Axis

```python
#create more fake data
other_data = data - (data % 150) + 50

#create the plot
fig, ax = plt.subplots()

#plot both sets of data
ax.plot(years,data)
ax.plot(years,other_data)

#add labels for the x and y axis
ax.set_xlabel("X Axis label")
ax.set_ylabel("Y Axis label")

ax.set_title("PLOT TITLE")
ax.legend(["Sample Data, Other Data"],loc =4)
```

```python
# Create fake data
x1 = [1, 4, 6, 8]
y1 = [10, 15, 27, 32]
x2 = [0.5, 2.2, 4.2, 6.5]
y2 = [21, 19, 9, 26]

# Create the plot
fig, ax = plt.subplots()

# Generate a line plot
ax.plot(x1, y1)

# Draw a scatter plot on same axes
ax.scatter(x2, y2)

# Add a legend
ax.legend(["Dataset 1", "Dataset 2"]);
```
