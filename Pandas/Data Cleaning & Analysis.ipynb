```python
df.info()
```

```python
# Line graph
# Create a new column 'TotalSales' by multiplying 'Quantity' and 'UnitPrice'
df['TotalSales'] = df['Quantity'] * df['UnitPrice']

# Extract year and month from InvoiceDate
df['Year'] = df['InvoiceDate'].dt.year
df['Month'] = df['InvoiceDate'].dt.month

# Group data by Year and Month and sum the 'TotalSales' for each group
monthly_sales = df.groupby(['Year', 'Month'])['TotalSales'].sum().reset_index()

# Display the resulting DataFrame
print(monthly_sales)
```

```python
# monthly sales line graph
plt.figure(figsize=(10, 6))
sns.lineplot(data=monthly_sales, x='Month', y='TotalSales', color='blue')
plt.xlabel('Month')
plt.ylabel('Number of Sales')
plt.title('Monthly Sales Trends Over Time')
plt.show()
```

```python
# 2. Correlation matrix
# Select the numerical variables for which you want to calculate the correlation matrix
numerical_vars = df[['Quantity', 'UnitPrice', 'TotalSales', 'Year', 'Month']]

#Calculate the correlation matrix using the corr() method
correlation_matrix = numerical_vars.corr()

#Visualize the correlation matrix using a heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True)
plt.title('Correlation Matrix')
plt.show()

```

There is positive correlation between Quantity and Totalsales, meaning as the quantity increases the total sales also increases.

5.3 Multivariate Analysis

```python
# Step 1: Descriptive Statistics
# Compute basic statistics for numerical variables
numerical_stats = df[['Quantity', 'UnitPrice', 'TotalSales', 'Year', 'Month']].describe()
print(numerical_stats)
```

Since there is only linear relation between two variables( quantities and total sales) then there is no need for us to do any dimensionality reduction. that is the PCA/LDA

### pca

```python
#this is how you would have done PCA
#PCA
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA#

#Step 1: Standardize the numerical variables
# PCA is sensitive to the scale of variables, so we standardize them.
scaler = StandardScaler()
scaled_data = scaler.fit_transform(df[['Quantity', 'UnitPrice', 'TotalSales', 'Year', 'Month']])
```

```python
# Step 2: Perform PCA
pca = PCA()
pca.fit(scaled_data)
```

```python
# Step 3: Explained Variance
# Check how much variance each principal component explains
explained_variance = pca.explained_variance_ratio_
print("Explained Variance Ratios:")
print(explained_variance)
```

```python
# Step 4: Select the number of components.
# You can decide how many principal components to keep based on explained variance.
# For example, if you want to keep 95% of the variance, you can use np.cumsum(explained_variance).
# Here, we'll keep all components for demonstration purposes.
```
