# Accessing Data within Pandas

```python
import pandas as pd
```

```python
# import data from the scikit-learn library

from sklearn.datasets import load_wine

data = load_wine()
df = pd.DataFrame(data.data, columns=data.feature_names)
```

```python
#reading the first 10 entries in the data
df.head(10)
```

These methods include the .head() that extracts the top attributes and .tail() that extracts the last attributes and the .info() that extracts a concise summary of the dataframe

```python
type(df)
```

```python
#the tail end of the data the last 5 rows

df.tail(5)
```

```python
# the info of the dataframe
df.info()
```

```python
# Using .index, you can access the index or row labels of the DataFrame

df.index
```

```python
df.columns  # this accesses the columns of the data
```

### Selecting the DataFrame information
The attributes for this are :
-    .iloc, this is used to locate the row of the data based on the index given
-    .loc, this has 2 use cases:
-                  Selecting by label/index
                   Selecting with boolean / conditional look up
              

```python
# iloc

df.iloc[3]
```

```python
# in cases where you want to select multiple rows u an use colons for that

df.iloc[1:8]  #similar to how it is done in lists
```

```python
# you can use the , to perform column selection based on index as well

df.iloc[:,3:7]
```

```python
# given this information you can perform row and column seection at the same time

df.iloc[5:10,3:9]
```

### `.loc`

```python
# label based indexing

df.loc[:,"magnesium"]
```

```python
df.loc[7:16,["magnesium","total_phenols"]]
```

### boolean indexing using the .loc

in the case below imagine you'd like to create a new DataFrame that only contains the wines with an alcohol percentage below 12. This can be done as follows:

```python

df.loc[df['alcohol'] < 12]
```

```python
# the above can be done using :
df[df['alcohol'] < 12]
```

```python
df.loc[df["alcohol"]<12,['color_intensity']]
```

### Selectors for series


```python
#creating series from one column dataframes 

col_intensity = df["color_intensity"]

type(col_intensity)
```

```python
col_intensity[:5]
```

```python
col_intensity[col_intensity >= 8]
```

### Changing the values and setting values in DataFrames and Series

```python

# Changing the values in instances where we want to set values for a certain reason
df.loc[df["color_intensity"] > 10 , 'color_intensity'] = 10
# df.loc[df["color_intensity"] >10]
df["color_intensity"]
```

```python
#creating new columns in this case a column called shade

df.loc[df["color_intensity"] > 7, "shade"] = "dark"
df.loc[df["color_intensity"] <= 7, "shade"] = "light"

df["shade"]
```

```python
df[df["shade"] == "dark"]
```

```python
df.shape # the columns have increased by one since we have created the shade column
```
