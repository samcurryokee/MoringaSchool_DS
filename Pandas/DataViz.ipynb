# Data Visualization with Pandas

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams["figure.figsize"] = (10,10)
plt.style.available
```

```python
#using a style
plt.style.use('ggplot')
```

### Dataset for the visualization

```python
import pandas as pd
import numpy as np

np.random.seed(777)

data = pd.DataFrame({'A':np.random.randn(365).cumsum(),
                    'B':np.random.randn(365).cumsum() + 25,
                    'C':np.random.randn(365).cumsum() - 25}, 
                     index = pd.date_range('1/1/2018', periods = 365))
data.head()
```

```python
data.plot()  #here we get a line graph
```

```python
#dreawing any other type of plot we use the kind parameter
data.plot("A","B",kind="scatter")
```

```python
ax = data.plot.scatter('A', 'C', 
                  c = 'B',
                  s = data['B'],
                  colormap = 'viridis')
ax.set_aspect("equal")
ax.set_title('Manipulating Pandas plot objects in matplotlib')
```

```python
data.plot.box()
```

```python
#Histograms
#setting alpha level to inspect the distribution level
data.plot.hist(alpha = 0.7);
```
