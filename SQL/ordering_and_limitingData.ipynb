# Ordering and Limiting Data with SQL

```python
import pandas as pd
import sqlite3
conn = sqlite3.connect("data.sqlite")
```

```python
pd.read_sql("""
SELECT * FROM products;
""",conn)
```

### `ORDER BY`

```python
#order products by the product name

pd.read_sql("""
SELECT * FROM products
ORDER BY productName;
""",conn)
```

when using the `ORDER BY` the default order is ascending order. Youo can also specify `ASC` at the end of the order by clause


```python
pd.read_sql("""
SELECT * FROM products
ORDER BY productName ASC;
""",conn)
```

```python
#to order this by descending order we can replace the ASC to DESC

pd.read_sql("""
SELECT * FROM products
ORDER BY productName DESC;
""",conn)
```

### Custom Sorting

Not limmited by the alphabetical order we can be able to sort the data we wan to retrieve

```python
#Sorting by  length of productDescription
pd.read_sql(""" 
SELECT productName,length(productDescription) AS description_length
FROM PRODUCTS
ORDER BY description_length;
""",conn)
```

```python
#The above can be written as follows but does not show the description length

pd.read_sql("""
SELECT productName
FROM products
ORDER BY length(productDescription);
""",conn)
```

### Sorting By Multiple Columns

```python
#sorts by productVendor then productName
pd.read_sql("""
SELECT productVendor ,productName, MSRP
FROM products
ORDER BY productVendor,productName;
""",conn)
```

```python
#sorts by productVendor then productName
pd.read_sql("""
SELECT productVendor ,productName, MSRP
FROM products
ORDER BY productName,productVendor;
""",conn)
```

since the above column the order has been changed so the output also changes.
The way to know which will come first is by checking which column has fewer unique values  ansd if so it should appear earlier

```python
pd.read_sql("""
SELECT COUNT(DISTINCT productVendor) AS num_product_vendors,
       COUNT(DISTINCT productName) AS num_pruduct_names
  FROM products;
""",conn)
```

since the productName has more unique values as compared to the productVendors then the productVendors appears 1st then names

### Type Casting

```python
#see the below example

pd.read_sql("""
SELECT productName , quantityInStock
FROM products
ORDER BY quantityInStock
""",conn).head(10)
```

Type cating is used in case where the output provided is is odd or not particularly right

In this case we assume that the quantity in stock is stored as integer or numerical value but in this case it is stored as a string

WE want to cast this as numbers not srings so we use the `CAST` as part of the `ORDER BY` clause

```python
pd.read_sql("""
SELECT productName,quantityInStock
FROM products
ORDER BY CAST(quantityInStock AS INTEGER);
""",conn).head()
```

### LIMIT

This is used to determine the number of records you want to return from a dataset

```python
pd.read_sql("""
SELECT * FROM orders
LIMIT 25;
""",conn)
```

You can use the `LIMIT` with the `ORDER BY` where it becomes more porwerful as a tool for filtering

```python
pd.read_sql("""
SELECT *
  FROM orders
 ORDER BY length(comments) DESC
 LIMIT 10;
""", conn)
```

You can combine LIMIT with WHERE

```python
pd.read_sql("""
SELECT *
  FROM orders
 WHERE status = "Cancelled"
 ORDER BY length(comments) DESC
 LIMIT 10;
""", conn)
```

```python
pd.read_sql("""
SELECT *,
       julianday(shippedDate) - julianday(orderDate) AS days_to_fulfill
  FROM orders
 WHERE shippedDate != ""
 ORDER BY days_to_fulfill DESC
 LIMIT 10;
""", conn)
```
